```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Titanic dataset
titanic = sns.load_dataset("titanic")
```

```python
#Daniel Justine L. Corcino
#BSCPE 3-2
#Big Data Analytics


# Style used for the plots
sns.set_style("darkgrid")

# Average Fare by Class and Gender using Bar plot
plt.figure(figsize=(8, 5))
sns.barplot(x="class", y="fare", hue="sex", data=titanic, palette=["#FF6666", "#6699FF"], edgecolor="black")
plt.xlabel("Class", fontsize=12)
plt.ylabel("Average Fare ($)", fontsize=12)
plt.title("Average Fare by Class and Gender", fontsize=12, fontweight="bold")
plt.legend(title="Gender")
plt.show()

# Age vs. Fare with survival status using Scatter plot
plt.figure(figsize=(8, 5))
sns.scatterplot(x="age", y="fare", data=titanic, hue="survived", palette=["#FF6666", "#66B2FF"], alpha=0.7, edgecolor="black")
plt.xlabel("Age", fontsize=12)
plt.ylabel("Fare ($)", fontsize=12)
plt.title("Age vs Fare with Survival Status", fontsize=12, fontweight="bold")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()

# Correlation Heatmap using Heatmap
plt.figure(figsize=(10, 6))
corr_matrix = titanic.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt='.2f', linewidths=0.5)
plt.title("Correlation Heatmap", fontsize=12, fontweight="bold")
plt.xticks(rotation=45, ha="right")
plt.yticks(rotation=0)
plt.show()
```
